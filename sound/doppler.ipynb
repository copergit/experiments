#  Studio dell'andamento dello spostamento Doppler per sorgenti in moto non lungo la linea di vista

Se l'osservatore non giace lungo la traiettoria dell'emettitore, sarà necessario determinare la componente della velocità dell'emettitore, lungo la linea di vista dell'osservatore. Facendo riferimento al grafico sottto riportato, consideriamo come legge oraria dell'emettitore, quella di un semplice moto a velocità costante, con origine in $\Omega$:

\begin{equation}
x(t)=vt
\end{equation}

detta $d$ la distanza dell'ossevatore dalla traiettoria dell'emettitore, ed $l$ la distanza del punto in cui la perpendicolare per l'osservatore interseca la traiettora dell'emettitore, possiamo facilmente scrivere il coseno dell'angolo $\vartheta$ riportato in figura:

\begin{equation}
\cos\vartheta=\frac{l-vt}{\sqrt{d^2+(l-vt)^2}}
\end{equation}



![caption](files/non_colinear_motion.svg)


possiamo provare a rappresentare l'andamento di questa funzione nel tempo e per vari valori dei parametri del sistema. Definiamo quindi una funzione:

```python
import math
def gamma(t,d,l,v):
    return  (l-v*t)/math.sqrt(pow(d,2)+pow((l-v*t),2))
```

```python
gamma(1,10,100,13)
```

```python
import matplotlib.pyplot as plt 

%matplotlib widget
plt.rcParams['animation.html']='jshtml'
```

```python
scale=0.03
time=[t*scale for t in range(0, 1000,1)]
```

```python
fig=plt.figure(figsize=[10,10])
plt.plot([t for t in time],[gamma(t,10,100,13) for t in time],'-',color='gray',markerfacecolor='red')
```

Calcoliamo ora la frequenza udita dall'osservatore che stiamo studiando

\begin{equation}
\nu'=\dfrac{v_s}{v_s-v\cos\gamma}\nu
\end{equation}

Scriviamo quindi la funzione che ci permette di calcolare la frequenza udita rispetto alla frequenza emessa in funzione della posizione dell'emettitore

```python
def freq(f_e,t,l,d,v):
    vs=340
    return vs/(vs-v*gamma(t,d,l,v))*f_e
```

```python
scale=0.3
time=[t*scale for t in range(0, 1000,1)]
fig=plt.figure(figsize=[10,10])
fe=100
plt.plot([t for t in time],[freq(fe,t,100,50,1) for t in time],'-',color='gray')
plt.plot([t for t in time],[fe for t in time],'-',color='red')
```

```python
scale=0.3
time=[t*scale for t in range(0, 1000,1)]
fig=plt.figure(figsize=[10,10])
fe=440
plt.plot([t for t in time],[freq(fe,t,100,0,1) for t in time],'-',color='gray')
plt.plot([t for t in time],[fe for t in time],'-',color='red')
```
